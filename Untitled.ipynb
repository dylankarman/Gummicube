# Introduction

Sample App can be found on the iOS App Store. Sample App made a change to a marketing campaign that impacted their App Store Search channel on 10/28/20. The developer of Sample App would like to know the following: What is the impact of the 10/28/20 store change? 

# Data Processing 

```python
# importing necessary packages and importing the csv file.
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv('SampleApp (2).csv')
```

```python
# taking a look at the dataframe to see if it imported properly.
df.head()
```

```python
# it seems that the first three rows are not needed. i will delete them.
df = df.drop(labels=[0,1,2], axis=0)
```

```python
# checking to see if dropping the unneeded rows worked.
df.head()
```

```python
# changing the name of the columns to the names in the first row.
df = df.rename(columns=df.iloc[0])
df.head()
```

```python
# great. now we'll delete the first row and last column because they are not needed....
df = df.drop(labels=[3], axis=0)
df.head()
```

```python
# and reset the index numbers so that our dataframe is ready to go.
df = df.reset_index(drop=True)
df.head()
```

```python
# looking at the column names.
df.columns
```

```python
# making a new dataframe without the column nan.
df = df[['Date', 'App Referrer Impressions','App Store Browse Impressions', 'App Store Search Impressions', 'Unavailable Impressions',     'Web Referrer Impressions',
'App Referrer App Units', 'App Store Browse App Units', 'App Store Search App Units', 'Unavailable App Units',
'Web Referrer App Units']]
```

```python
# checking to see if it worked.
df.head()
```

```python
# making sure that the date is in the correct format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# checking the dataframe to see if there are any missing values or something that catches my eye in the data.
# there are no missing values. what a relief. 
df.info()
```

```python
# there are numerical values expressed as objects so I changed them to integers.
df['App Referrer Impressions'] = df['App Referrer Impressions'].astype(int)
df['App Store Browse Impressions'] = df['App Store Browse Impressions'].astype(int)
df['App Store Search Impressions'] = df['App Store Search Impressions'].astype(int)
df['Unavailable Impressions'] = df['Unavailable Impressions'].astype(int)
df['Web Referrer Impressions'] = df['Web Referrer Impressions'].astype(int)
df['App Referrer App Units'] = df['App Referrer App Units'].astype(int)
df['App Store Browse App Units'] = df['App Store Browse App Units'].astype(int)
df['App Store Search App Units'] = df['App Store Search App Units'].astype(int)
df['Unavailable App Units'] = df['Unavailable App Units'].astype(int)
df['Web Referrer App Units'] = df['Web Referrer App Units'].astype(int)
```
